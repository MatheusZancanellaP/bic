```python

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('global_irradiance_daily_aurora.csv', parse_dates=['Data'], delimiter=';')
df['Medicao'] = pd.to_numeric(df['Medicao'], errors='coerce')


total_dias = len(df)
dias_sem_medicao = df['Medicao'].isna().sum()


porcentagem_sem_medicao = (dias_sem_medicao / total_dias) * 100
print(f"Porcentagem de dias sem medição: {porcentagem_sem_medicao:.2f}%")


plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['Medicao'])
plt.title('Série Temporal de Medição de Irradiância Global')
plt.xlabel('Data')
plt.ylabel('Medição')
plt.grid(True)
plt.show()

```

```python
df.sort_values('Data', inplace=True)


df['Sem_Medicao'] = df['Medicao'].isna()


mudanca_indices = np.where(df['Sem_Medicao'].diff())[0]


intervalos = np.diff(mudanca_indices)


maior_intervalo = max(intervalos) if len(intervalos) > 0 else 0


print(f"O maior intervalo de dias com medições é de {maior_intervalo} dias.")
```

```python
if len(intervalos) > 0:
    maior_intervalo = max(intervalos)
    indice_max = np.argmax(intervalos)
    inicio_intervalo = mudanca_indices[indice_max]
    fim_intervalo = mudanca_indices[indice_max + 1]
    
   
    df_intervalo = df.iloc[inicio_intervalo:fim_intervalo]
    
   
    df_intervalo.to_csv('maior_intervalo_com_medicoes.csv', index=False, columns=['Data', 'Medicao'])
    
    print(f"O maior intervalo de dias com medições é de {maior_intervalo} dias.")
    print(f"Os dados desse intervalo foram salvos em 'maior_intervalo_com_medicoes.csv'")
    plt.figure(figsize=(10, 6))
    plt.plot(df_intervalo['Data'], df_intervalo['Medicao'])
    plt.title('Série Temporal do Maior Intervalo com Medições')
    plt.xlabel('Data')
    plt.ylabel('Medição')
    plt.grid(True)
    plt.show()
```

```python
def introduce_gaps(data, missing_percentage=0.2, random_seed=42):
    np.random.seed(random_seed)
    data_with_gaps = data.copy()
    num_missing = int(missing_percentage * len(data))
    missing_index = np.random.choice(data.index, size=num_missing, replace=False)
    data_with_gaps.loc[missing_index, 'Medicao'] = np.nan  
    return data_with_gaps


df_intervalo_with_gaps = introduce_gaps(df_intervalo, missing_percentage=0.2)


df_intervalo_with_gaps.to_csv('df_intervalo_with_gaps.csv', index=False)
```

```python
df_intervalo_with_gaps['Medicao'].interpolate(method="linear", limit_direction="backward", inplace=True)


df_intervalo_with_gaps.to_csv("df_interpolate.csv", index=False)

y_true = df_intervalo['Medicao'].values
y_pred = df_intervalo_with_gaps['Medicao'].values


print("R2:", r2_score(y_true, y_pred))
print("MSE:", mean_squared_error(y_true, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
print("MAE:", mean_absolute_error(y_true, y_pred))
print("MAPE:", np.mean(np.abs((y_true - y_pred) / y_true)) * 100)
```

```python
# Plotar ACF

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure()
plot_acf(df_intervalo['Medicao'], lags=20)
plt.title('Autocorrelation Function (ACF)')
plt.show()

# Plotar PACF
plt.figure()
plot_pacf(df_intervalo['Medicao'], lags=20)
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df_intervalo['Medicao'])


print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Number of lags used:', result[2])
print('Number of observations used:', result[3])
print('Critical Values:', result[4])

# Interpretar o resultado com base no valor p
if result[1] <= 0.05:
    print("A série é estacionária.")
else:
    print("A série é não-estacionária e pode precisar ser diferenciada.")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

# 1. Preparação dos Dados

# Supondo que 'medição' seja sua coluna de interesse
values = df['medição'].values.reshape(-1, 1)

# Normalizar os dados
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)

# Transformar em sequências (por exemplo, sequências de tamanho 10)
X, y = [], []
for i in range(len(scaled) - 10):
    X.append(scaled[i:i+10])
    y.append(scaled[i+10])
X, y = np.array(X), np.array(y)

# 2. Construção da LSTM

def create_model(lstm_units=50, dropout=0.2):
    model = Sequential()
    model.add(LSTM(lstm_units, input_shape=(X.shape[1], X.shape[2]), dropout=dropout))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam')
    return model

# 3. GridSearch

model = KerasRegressor(build_fn=create_model, epochs=30, batch_size=32, verbose=0)

# Definir os hiperparâmetros para teste
param_grid = {
    'lstm_units': [30, 50, 80],
    'dropout': [0.1, 0.2, 0.3],
    'batch_size': [16, 32, 64],
    'epochs': [10, 30, 50]
}

grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)
grid_result = grid.fit(X, y)

# 4. Plotar os Resultados

results = pd.DataFrame(grid_result.cv_results_)
plt.plot(results['param_lstm_units'], results['mean_test_score'])
plt.xlabel('LSTM Units')
plt.ylabel('Negative MSE')
plt.show()
```
